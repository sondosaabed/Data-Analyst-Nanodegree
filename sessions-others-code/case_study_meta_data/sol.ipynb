## Loading the Datsets

```python
import pandas as pd
import numpy as np

daily_activity_data = pd.read_csv("./case_study/dailyActivity_merged.csv")
daily_calories_data = pd.read_csv("./case_study/dailyCalories_merged.csv")
daily_steps_data = pd.read_csv("./case_study/dailySteps_merged.csv")
sleep_day_data = pd.read_csv("./case_study/sleepDay_merged.csv")
weight_log_data = pd.read_csv("./case_study/weightLogInfo_merged.csv")
```
