```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get('https://www.maannews.net')
soup= BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
for link in soup.find_all('a', class_='list-3__item default'):
    print(link.get('href'))
```

```python
for link in soup.find_all('a', class_='list-3__item default'):
    print(link.get('href'), link.getText())
```

```python
df=pd.DataFrame(columns=['title', 'url'])
```

```python
df['title']=[link.getText() for link in soup.find_all('a', class_='list-3__item default')]
```

```python
df['url']=[link.get('href') for link in soup.find_all('a', class_='list-3__item default')]
```
